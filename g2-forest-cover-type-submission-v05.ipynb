# Forest Cover Type

# Import Statements

```python
import pandas as pd, numpy as np
import joblib
```

# Load Data

```python
# Load the test data set
pd.set_option('max_columns', None)
test = pd.read_csv('../input/forest-cover-type-prediction/test.csv')
# Print the shape of the test set 
print(test.shape)
```

```python
# view 
test.head()
```

```python
# remove ID column from set
test = test.iloc[:, 1:]
test.head()
```

# Check for Missing Values

```python
# check for missing values
test.isnull().values.any()
```

# Preprocessing

```python
# create cat and num variables
X_cat = test.iloc[:, 10:54].values
X_num = test.iloc[:, 0:10].values

print(f'Categorical Shape: {X_cat.shape}')
print(f'Numerical Shape: {X_num.shape}')
```

```python
# load scaler
scaler = joblib.load('../input/forest-cover-models/forest_cover_scaler.joblib')
# apply scaler to test data
X_num = scaler.transform(X_num)
```

```python
# combine num and cat
X = np.hstack((X_num, X_cat))
print(X.shape)
```

# Model Predictions

```python
# load model
model = joblib.load('../input/forest-cover-models/ensemble_model.joblib')
```

```python
# make predictions using model
y_pred = model.predict(X)
```

# Submission

```python
# write to submission
submission = pd.read_csv('../input/forest-cover-type-prediction/sampleSubmission.csv')
# view first few rows
submission.head()
```

```python
# add predictions to file
submission['Cover_Type'] = y_pred
# view first few rows
submission.head()
```

```python
# export to csv file
submission.to_csv('submission.csv', index = False, header = True)
```
